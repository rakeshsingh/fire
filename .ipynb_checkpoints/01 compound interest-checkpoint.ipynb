### Compound interest as the eighth wonder of the world


Compound interest is the eighth wonder of the world. He who understands it, earns it ... he who doesn't ... pays it. — Albert Einstein


```python
import pandas as pd
import datetime
pd.set_option('display.width', 200) # Displaying more columns in one row

# Data date range, Google provides up to 4000 entries in one call
start = datetime.datetime(2000, 1, 1) 
start_2 = datetime.datetime(2000, 1, 1)
end = datetime.datetime(2024, 1, 1)

# compound interest seeds
interest = 8
initial = 1000
retirement_age = 50
months = 12*retirement_age

early_start_recurring = 100
## someone started 10yr late, and invests twice the amount
late_start_recurring = 200
early_start =[]
late_start = []

for i in range(0,240): 
    early_start.append(0)
early_start.append(initial)
for i in range(241,months):
    early_start.append( (early_start[i-1]* (1+ interest/12/100)) + early_start_recurring)


# late start calculation
for i in range(0,360): 
    late_start.append(0)
late_start.append(initial)
for i in range(361,months):
    late_start.append( (late_start[i-1]* (1+ interest/12/100)) + late_start_recurring)


```
